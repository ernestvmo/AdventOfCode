```python
def load_data(filename: str):
    with open(filename) as file:
        moves = [move.rstrip("\n").split() for move in file]
        moves = [[move[0], int(move[1])] for move in moves]
    return moves
```

```python
def rope_board(moves: list):
    L = R = D = U = 1
    for dir, dist in moves:
        if dir == "L":
            L += dist
        elif dir == "R":
            R += dist
        elif dir == "D":
            D += dist
        else:
            U += dist
    columns = R + L if (R + L) % 2 == 1 else R + L + 1
    rows = U + D if (U + D) % 2 == 1 else U + D + 1
    center = [U, L]
    print(columns, rows)
    return [["." for _ in range(columns)] for _ in range(rows)], center

def update_board(new_pos, head, tail, board):
    board[new_pos[0]][new_pos[1]] = "H"
    board[head[0]][head[1]] = "."
    if board[tail[0]][tail[1]] != "s":
        board[tail[0]][tail[1]] = "T"
    return board

def new_position(old_pos: list, move: list) -> list:
    new_pos = old_pos[:]
    if move[0] == "R":
        new_pos[1] += move[1]
    elif move[0] == "L":
        new_pos[1] -= move[1]
    elif move[0] == "D":
        new_pos[0] += move[1]
    elif move[0] == "U":
        new_pos[0] -= move[1]
    return new_pos

def is_adjacent(H, T):
    if abs(H[0] - T[0]) <= 1 and abs(H[1] - T[1]) <= 1:
        return True
    else:
        return False

def same_line(H, T):
    if H[0] == T[0] or H[1] == T[1]:
        return True
    else:
        return False

def move_diagonal(head, tail, move):
    if move[0] == "U" or move[0] == "D":
        tail = new_position(tail, move)
        tail[1] = head[1]
        return tail
    elif move[0] == "R" or move[0] == "L":
        tail = new_position(tail, move)
        tail[0] = head[0]
        return tail

def move_rope(head, tail, move, board):
    new_pos = new_position(head, move)
    if board[head[0]][head[1]] == "s":
        board = update_board(new_pos, head, tail, board)
        return new_pos, tail
    if is_adjacent(new_pos, tail):
        # print("ADJACENT")
        board = update_board(new_pos, head, tail, board)
        return new_pos, tail
    else:
        # print("NOT ADJACENT!!!")
        if same_line(new_pos, tail):
            # print("SAME LINE")
            tail = new_position(tail, move)
            board = update_board(new_pos, head, tail, board)
            return new_pos, tail
        else:
            # print("DIAGONAL")
            tail = move_diagonal(new_pos, tail, move)
            board = update_board(new_pos, head, tail, board)
            return new_pos, tail

def print_board(board, visited=None):
    if visited is not None:
        for i in range(len(board)):
            for j in range(len(board[i])):
                if [i, j] in visited:
                    board[i][j] = "#"
                else:
                    board[i][j] = "."
    for l in board:
        print(l)
    print() 

def move(moves: list):
    board, center = rope_board(moves)
    # board = [['.' for _ in range(100)] for _ in range(100)]
    print(center)
    board[center[0]][center[1]] = 's'
    # print_board(board)
    # center = [len(board) - 1, 0]
    head, tail = center[:], center[:]
    visited = []
    for move in moves:
        for _ in range(move[1]):
            head, tail = move_rope(head, tail, [move[0], 1], board)
            if tail not in visited:
                visited.append(tail)
            # print_board(board)
    print(len(visited))
    # print_board(board, visited)
```
